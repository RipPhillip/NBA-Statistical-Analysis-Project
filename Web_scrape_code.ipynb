```python
years = [str(i) for i in range(2005, 2021)]
teams = ["ATL", "BRK", "BOS", "CHO", "CHI", "CLE", "DAL", "DEN",
                "DET", "GSW", "HOU", "IND", "LAC", "LAL", "MEM", "MIA",
                "MIL", "MIN", "NOP", "NYK", "OKC", "ORL", "PHI", "PHO",
                "POR", "SAC", "SAS", "TOR", "UTA", "WAS", "NJN", "NOH", "CHA", "CHH", "SEA", "VAN"
                "NOK"]
```

```python
# pseudocode:    
# create a master dataframe(initially empty) with the headers

# URL where game log table is
url = "https://www.basketball-reference.com/teams/PHI/2014/gamelog/"

# this is the HTML from the given URL
html = urlopen(url)
soup = BeautifulSoup(html)

# use getText() to extract text of the table header (stat/column names)
basic_headers = [th.getText() for th in soup.findAll('thead', limit=2)[0].findAll('th')]
# keep desired stats/column names
basic_headers = basic_headers[7:30]

# URL where game log table is
url = "https://www.basketball-reference.com/teams/PHI/2014/gamelog-advanced/"

# this is the HTML from the given URL
html = urlopen(url)
soup = BeautifulSoup(html)

# use getText() to extract the text of the adv. table header (stat/column names)
adv_headers = [th.getText() for th in soup.findAll('thead', limit=2)[0].findAll('th')]
# keep desired stats/column names
adv_headers = adv_headers[16:]

# initialize master_df, will have header with no rows
master_df = pd.DataFrame(columns = (basic_headers + adv_headers + ['Team', 'Season']) )

# for each year (2010 - 2019, 10 seasons), get the teams' gamelogs

# for each team, get the basic and advanced game logs
for year in years:
############################## Teams Full Game Logs 2014 ################################
    for team in teams:
        # for the basic game log, remove opponent stats (by indexing)
        # NBA season we will be analyzing
        
    
        ##### basic metrics #####
        stat_type = "gamelog"
        # URL where game log table is
        url = "https://www.basketball-reference.com/teams/{}/{}/{}/".format(team, year, stat_type)

        # this is the HTML from the given URL
        html = urlopen(url)
        soup = BeautifulSoup(html)

        # begin at third table row (first game log)
        rows = soup.findAll('tr')[2:]
        # extract stats of desired columns. nested list of each game log (row)
        basic_gl_lists = [[td.getText() for td in rows[i].findAll('td')[:-17]]
                          for i in range(len(rows))]
        # create df for basic game logs
        df_basic = pd.DataFrame(basic_gl_lists, columns = basic_headers)

        ##### advanced metrics #####
        stat_type = "gamelog-advanced"
        # URL where adv. game log table is
        url = "https://www.basketball-reference.com/teams/{}/{}/{}/".format(team, year, stat_type)

        # this is the HTML from the given URL
        html = urlopen(url)
        soup = BeautifulSoup(html)

        # begin at third table row (first game log)
        rows = soup.findAll('tr')[2:]
        # extract stats of desired columns. nested list of each game log (row)
        adv_gl_lists = [[td.getText() for td in rows[i].findAll('td')[7:]]
                        for i in range(len(rows))]
        # create df for adv. game logs
        df_adv = pd.DataFrame(adv_gl_lists, columns = adv_headers)

        # merge the basic and advanced game logs side by side
        team_df = pd.concat([df_basic, df_adv], axis=1)
        
        # Add the team names to the DF
        team_df['Team'] = team
        
        # Add the season year to the DF
        team_df['Season'] = year

        # add the dataframe to the master dataframe
        master_df = pd.concat([master_df, team_df])
```
