```python
import pandas as pd
import numpy as np


df = pd.read_csv('data/unclean_data.csv').iloc[:,1:]
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
#Change the dates to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.columns
```

```python
#Change column name to Home/away
df.rename(columns={'\xa0':'Home', 'eFG%.1':'D-eFG%', 'TOV%.1':'D-TOV%', 'FT/FGA.1':'D-FT/FGA',\
                  'Tm':'PTS', 'Opp.1':'OppPTS', 'DRB%':'D-DRB%', 'W/L':'Win'}, inplace=True)
```

```python
# Drop the extra/pointless columns
df.drop(df.columns[[33, 38, 0]], axis=1, inplace=True)
```

```python
# Fill the home/away column with 1s and 0s
df['Home'] = df['Home'].fillna(1)
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.replace(to_replace = '@', value = '0', inplace=True)
```

```python
df.replace(to_replace = 'W', value = '1', inplace=True)
```

```python
df.replace(to_replace = 'L', value = '0', inplace=True)
```
