## Create copies of the advanced DFs in order to check which metrics are heavily impacting the models

```python
clipped_advanced_df = advanced_df.copy()
```

```python
# Create correlation matrix
corr_matrix = clipped_advanced_df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.0
to_drop = [column for column in upper.columns if any(upper[column] > 0.60)]

# Drop features 
clipped_advanced_df.drop(to_drop, axis=1, inplace=True)
```

## Build a bunch of vanilla classifiers with advanced stats

```python
clipped_advanced_df.columns
```

```python
# Split the data

y = advanced_df['Win']
X = advanced_df.drop(columns=['Win'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=42)
```

```python
classifiers = {
    'Logistic Regression' : LogisticRegression(), 
    'Random Forest' : RandomForestClassifier(), 
    'Support Vector Machine' : SVC(), 
    'K-Nearest Neighbors' : KNeighborsClassifier(), 
    'Decision Tree' : DecisionTreeClassifier(), 
    'XGBoost' : XGBClassifier()
}
```

```python
# Use a loop to go through each vanilla model
results = pd.DataFrame(columns= ['Train_accuracy', 'Test_accuracy', 'F1_score', 
                                'False_Negative', 'True_Positive'])

for key, value in classifiers.items():
    #fit models
    value.fit(X_train, y_train)
    train_pred = value.predict(X_train)
    y_pred = value.predict(X_test)
    
    # get accuracy, f1 score
    train_acc = accuracy_score(y_train, train_pred) * 100
    test_acc = accuracy_score(y_test, y_pred) * 100
    f1 = f1_score(y_test, y_pred)
    
    #get false neg and true positive
    cm = confusion_matrix(y_test, y_pred)
    FN = cm[1][0]
    TP = cm[1][1]
    
    # add measurements to datafram
    results.loc[key] = [round(train_acc, 2), round(test_acc, 2), 
                        round(f1, 2), round(FN, 0), round(TP, 0)]
```

```python
results.sort_values(by= ['F1_score', 'False_Negative'], ascending= False)

plt.savefig("One_team_results.png")
```

## Build a function to get results after each model

```python
def get_results(model, model_name):
    train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    # get classification report
    print('{} Classification Report'.format(model_name))
    print(classification_report(y_test, y_pred))
    
    # get confusion matrix
    plot_confusion_matrix(model, X_test, y_test, cmap= "Blues", values_format= '.5g')
    plt.grid(False)
    plt.show()
    
    # get accuracy and F1 scores
    train_acc = accuracy_score(y_train, train_pred) * 100
    test_acc = accuracy_score(y_test, y_pred) * 100
    f1 = f1_score(y_test, y_pred)
    
    #get false neg and true positive
    cm = confusion_matrix(y_test, y_pred)
    FN = cm[1][0]
    TP = cm[1][1]
    
    # save measurements into results df
    results.loc[model_name] = [round(train_acc, 2), round(test_acc, 2), 
                               round(f1, 2), round(FN, 0), round(TP, 0)]
```

## Vanilla KNN Classifier

```python
# Split the data

y = advanced_df['Win']
X = advanced_df.drop(columns=['Win', 'eFG%', 'ORtg'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=42)
```

```python
# Instantiate StandardScaler
scaler = StandardScaler()

# Transform the training and test sets
scaled_data_train = scaler.fit_transform(X_train)
scaled_data_test = scaler.transform(X_test)
```

```python
# Instantiate KNeighborsClassifier
clf = KNeighborsClassifier()

# Fit the classifier
clf.fit(scaled_data_train, y_train)

# Predict on the test set
test_preds = clf.predict(scaled_data_test)
```

```python
def print_metrics(labels, preds):
    print("Precision Score: {}".format(precision_score(labels, preds)))
    print("Recall Score: {}".format(recall_score(labels, preds)))
    print("Accuracy Score: {}".format(accuracy_score(labels, preds)))
    print("F1 Score: {}".format(f1_score(labels, preds)))
    
print_metrics(y_test, test_preds)
```

```python
# Transform/normalize data

clf = KNeighborsClassifier()
clf.fit(scaled_data_train, y_train)
test_preds = clf.predict(scaled_data_test)

def find_best_k(X_train, y_train, X_test, y_test, min_k=1, max_k=25):
    best_k = 0
    best_score = 0.0
    for k in range(min_k, max_k+1, 2):
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        preds = knn.predict(X_test)
        f1 = f1_score(y_test, preds)
        if f1 > best_score:
            best_k = k
            best_score = f1
    print("Best Value for k: {}".format(best_k))
    print("F1-Score: {}".format(best_score))
    
find_best_k(scaled_data_train, y_train, scaled_data_test, y_test)
```

## Decision Tree

```python
# Create features and labels
y = advanced_df['Win']
X = advanced_df.drop(columns = ['Win', 'eFG%', 'ORtg'], axis=1)

# Split the data, 25/75
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=42)

# Train Decision Tree classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

#Make predictions for test data
y_preds = clf.predict(X_test)
```

```python
# Calculate accuracy 
acc = accuracy_score(y_test, y_preds) * 100
print('Accuracy is :{0}'.format(acc))

# Check the AUC for predictions
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test,\
                                                               y_preds)
roc_auc = auc(false_positive_rate, true_positive_rate)
print('\nAUC is :{0}'.format(round(roc_auc, 2)))

# Create and print a confusion matrix 
print('\nConfusion Matrix')
print('----------------')
pd.crosstab(y_test, y_preds, rownames=['True'], colnames=['Predicted'],\
            margins=True)
```

```python
# Instantiate and fit a DecisionTreeClassifier
clf_2 = DecisionTreeClassifier(random_state=42, criterion='entropy', max_depth=5)
clf_2.fit(X_train, y_train)

# Plot and show decision tree
plt.figure(figsize=(12,12), dpi=500)
tree.plot_tree(clf_2, 
               feature_names=X.columns,
               class_names=np.unique(y).astype('str'),
               filled=True, rounded=True)
plt.show()
```

```python
get_results(clf_2, 'Decision Tree')
```

## Random Forest

```python
# Instantiate a classifier
rando = RandomForestClassifier()

# Create a parameter grid
param_grid_rando = {
                 'n_estimators': [10,30,100],
                 'criterion':['gini', 'entropy'],
                 'max_depth': [None, 2, 6, 10],
                 'min_samples_split': [5, 10],
                 'min_samples_leaf': [3,6]
}
```

```python
# Instantiate Randomized Search
rando_search = RandomizedSearchCV(rando, param_grid_rando, n_iter= 20, scoring= 'f1', 
                           random_state= 42)

# Fit the randomized search model
rando_search.fit(X_train, y_train)
```

```python
# Find best parameters for the Random Forest
rando_search.best_params_
```

```python
# See model results
get_results(rando_search, 'Random Search Random Forest')
```
